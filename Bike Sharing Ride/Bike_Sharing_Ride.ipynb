# Using hour DataSet to predict cnt(count of casual + registered users )

for more about the features in dataset kindly see the readme.txt file in the Data folder.

## Data Prepossing (One-Hot Encoding)

```python
season = pd.get_dummies(hour['season'], prefix='season')
season.head()
```

```python
month = pd.get_dummies(hour['mnth'], prefix='month')
month.head()
```

```python
hr = pd.get_dummies(hour['hr'], prefix='hr')
hr.head()
```

```python
weekday = pd.get_dummies(hour['weekday'], prefix='weekday')
weekday.head()
```

```python
weathersit = pd. get_dummies(hour['weathersit'], prefix='weathersit')
weathersit.head()
```

# #Concatinating all dummies to the dataset and droping old features

 Note : We can concatinate and drop the features at once i will show you that on days dataset for now i am doing
       them separetly.

```python
hour = pd.concat([hour, season, month, hr, weekday, weathersit], axis=1)
hour.head()
```

Now droping the actual features.

```python
hour = hour.drop(['instant','season', 'mnth','hr', 'weekday', 'weathersit'], axis=1)
hour.head()
```

Converting date to datatype date

```python
hour['dteday'] = pd.to_datetime(hour.dteday)
```

## Standarizeing features

```python
for field in ['casual', 'registered', 'cnt']:
    mean, std = hour[field].mean(), hour[field].std()
    hour.loc[:,field] = (hour[field]-mean)/std
```

```python
hour['casual'].head()
```

```python
train = hour.iloc[0:1390,:]
test = hour.iloc[1390:,:]
```
